```python
# Tinting : mixing the image colours with white

def tint(img,percent):
    tinted_img = img + (np.ones(img.shape)-img)*percent
    return(tinted_img)

ouli = plt.imread('ouli.png')
tinted_ouli = tint(ouli,0.5)

plt.imshow(tinted_ouli)
```
