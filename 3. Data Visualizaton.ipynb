<b><li> Multiple Subplots

```python
# Matplotlib hierarchy: A figure contains 1 axes (subplots) or more
# the axe (subplot) contains: 
     # axis (x and y)
     # spines (4 frames)
     # ticks

# Creating a figure and a set of subplots (axes):
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,sharey=True)

# creating the data points to plot
x = [1,2,3,4,5]
y1 = [2*num for num in x]
y2 = [-2*num for num in x]

# plotting the data points on the subplots
ax1.plot(x,y1,'--b',linewidth=2.5) # define both style and color
ax2.plot(x,y2,linestyle='-.',color='red',linewidth=5) # define the style and color separately

# setting the labels and titles
ax1.set(xlabel='x axis',ylabel='y axis',title='subplot1')
ax2.set(xlabel='x axis',ylabel='y axis',title='subplot2')

# adding text to the axis:
ax1.text(3,0,'left',ha='center',size=20,color='blue')
ax2.text(3,0,'right',ha='center',size=20,color='red')

plt.show()
```

```python
fig, ax = plt.subplots(3,3,sharey=True,sharex=True,facecolor='lavender')

color = np.array([['aquamarine','turquoise','lightseagreen'],
                  ['thistle','plum','violet'],
                  ['lemonchiffon','khaki','palegoldenrod']])

for row in range(3):
    for col in range(3):
        ax[row,col].text(0.5,0.5,str((row,col)),ha='center')
        ax[row,col].set_xticks(())
        ax[row,col].set_yticks(())
        ax[row,col].set_facecolor(color[row,col])
```

<b><li> Multiple lines in one plot + Axis ranges

```python
days = [1,2,3,4,5,6,7]
min_temp = [3,0,2,-1,-2,4,5]
max_temp = [20,18,19,17,22,23,23]

fig, ax = plt.subplots()

ax.plot(days,min_temp,'c',
        days,min_temp,'ob',
        days,max_temp,'y',
        days,max_temp,'or')

ax.set(xlabel='Days',ylabel='Temperature in Celsius',title='Daily Min and Max Temperatures')

# set our own axis range:
# we can show the existing axis range through: ax.axis()
ax.axis([0.5,7.5,-5,25])

plt.show()
```

<b><li> Linspace to create x values + Shading regions

```python
# we can use the numpy function linspace to create x values
x = np.linspace(-2*np.pi, 2*np.pi, 60)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x,y1,x,y2)

# shading: filling the area between 2 lines (x,y1) and (x,y2) x,y1,y2
ax.fill_between(x,0,y1,color='blue',alpha=0.2,where=y1>0)
ax.fill_between(x,0,y1,color='red',alpha=0.2,where=y1<0)
```

<b><li> Spines and Ticks

```python
# customizing ticks
fig, ax = plt.subplots()

# set ticks location
ax.set_xticks([1,3,5,7,9,10])
ax.set_yticks([0,2,4,6,8,10])

# set ticks labels
ax.set_xticklabels(['Japan','Korea','Thailand','China','Cambodia'])

# set ticks position
ax.yaxis.set_ticks_position('both')

# increase the font size of the tick labels and draw a box around them:
for xtick in ax.get_xticklabels():
    xtick.set_fontsize(13)
    xtick.set_bbox(dict(facecolor='green',edgecolor='black',alpha=0.5))
```

```python
# customizing spines
x = np.linspace(-2*np.pi, 2*np.pi, 60)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x,y)

# making the top and right spines invisible:
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# moving the bottom and left spines to 0 position
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# set the locator of major ticks
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))

# set ticks label
ax.set_xticklabels([r'$-2\pi',r'$-\frac{3\pi}{2}$',r'$-\pi$',r'$-\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', 
                    r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$'])
```

<b><li> Legends and Annotations

```python
# Legends
X = np.linspace(0,100,1000)
Y = [(x*x+2*x+1) for x in X]
Z = [(-x*x+2*x+1) for x in X]

fig, ax = plt.subplots()

# make sure to define the label (we use latex math)
label1 = r'$x^2-2x+1$'
label2 = r'$-x^2-2x+1$'
ax.plot(X,Y,label=label1)
ax.plot(X,Z,label=label2)

# set the legend location, othrwise it won't show up
#ax.legend(loc='upper left')
# if we don't know what the plot will look like we can let matplotlib decide the best one
ax.legend(loc='best')

# Annotations
ax.annotate("starting point",xy=(0,0),xytext=(0,4000),arrowprops=dict(facecolor='red'))
ax.annotate("ending point",(X[999],Y[999]),xytext=(79.5,5000),arrowprops=dict(facecolor='red'))
```
