<b><li> Multiple Subplots

```python
# Matplotlib hierarchy: A figure contains 1 axes (subplots) or more
# the axe (subplot) contains: 
     # axis (x and y)
     # spines (4 frames)
     # ticks

# Creating a figure and a set of subplots (axes):
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,sharey=True)

# creating the data points to plot
x = [1,2,3,4,5]
y1 = [2*num for num in x]
y2 = [-2*num for num in x]

# plotting the data points on the subplots
ax1.plot(x,y1,'--b',linewidth=2.5) # define both style and color
ax2.plot(x,y2,linestyle='-.',color='red',linewidth=5) # define the style and color separately

# setting the labels and titles
ax1.set(xlabel='x axis',ylabel='y axis',title='subplot1')
ax2.set(xlabel='x axis',ylabel='y axis',title='subplot2')

# adding text to the axis:
ax1.text(3,0,'left',ha='center',size=20,color='blue')
ax2.text(3,0,'right',ha='center',size=20,color='red')

plt.show()
```

```python
fig, ax = plt.subplots(3,3,sharey=True,sharex=True,facecolor='lavender')

color = np.array([['aquamarine','turquoise','lightseagreen'],
                  ['thistle','plum','violet'],
                  ['lemonchiffon','khaki','palegoldenrod']])

for row in range(3):
    for col in range(3):
        ax[row,col].text(0.5,0.5,str((row,col)),ha='center')
        #ax[row,col].set_xticks(())
        #ax[row,col].set_yticks(())
        ax[row,col].set_facecolor(color[row,col])
        
# sometimes the layout is tight and ticks cover each other, for more padding we use:
plt.tight_layout()
```

<b><li> Spines and Ticks

```python
# customizing ticks
fig, ax = plt.subplots()

# set ticks location
ax.set_xticks([1,3,5,7,9,10])
ax.set_yticks([0,2,4,6,8,10])

# set ticks labels
ax.set_xticklabels(['Japan','Korea','Thailand','China','Cambodia'])

# set ticks position
ax.yaxis.set_ticks_position('both')

# increase the font size of the tick labels and draw a box around them:
for xtick in ax.get_xticklabels():
    xtick.set_fontsize(13)
    xtick.set_bbox(dict(facecolor='green',edgecolor='black',alpha=0.5))
```

```python
# customizing spines
x = np.linspace(-2*np.pi, 2*np.pi, 60)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x,y)

# making the top and right spines invisible:
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# moving the bottom and left spines to 0 position
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# set the locator of major ticks
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))

# set ticks label
ax.set_xticklabels([r'$-2\pi',r'$-\frac{3\pi}{2}$',r'$-\pi$',r'$-\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', 
                    r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$'])

plt.show()
```

```python
# Colored ticklabels
fig, ax1 = plt.subplots()

X = np.arange(1,10,0.1)

ax1.plot(X,2*np.pi*X,'b')
ax1.set_ylabel('Circumference (cm)',color='b')
for label in ax1.get_yticklabels():
    label.set_color('blue')

# Create a twin axes sharing the x axis
ax2 = ax1.twinx()
ax2.plot(X,np.pi*X**2,'r')
ax2.set_ylabel('Area (cm2)',color='red')
for label in ax2.get_yticklabels():
    label.set_color('red')

# saving figures (PNG, JPG, EPS, SVG, PGF and PDF)
fig.savefig('myfig.png',dpi=200)
```

<b>add_axes</b>

The calling signature of add_axes is add_axes(rect), where rect is a list [x0, y0, width, height] denoting the lower left point of the new axes in figure coodinates (x0,y0) and its width and height. So the axes is positionned in absolute coordinates on the canvas. 

add_axes() as shown below places a figure in the canvas that is exactly as large as the canvas itself.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

<b>add_subplot</b>

The calling signature of add_subplot does not directly provide the option to place the axes at a predefined position. It rather allows to specify where the axes should be situated according to a subplot grid. The usual and easiest way to specify this position is the 3 integer notation.

In this example a new axes is created at the first position (1) on a grid of 2 rows and 3 columns. To produce only a single axes, add_subplot(111) would be used (First plot on a 1 by 1 subplot grid). (In newer matplotlib versions, add_subplot() without any arguments is possible as well.)

The advantage of this method is that matplotlib takes care of the exact positioning. By default add_subplot(111) would produce an axes positioned at [0.125,0.11,0.775,0.77] or similar, which already leaves enough space around the axes for the title and the (tick)labels. However, this position may also change depending on other elements in the plot, titles set, etc. It can also be adjusted using pyplot.subplots_adjust(...) or pyplot.tight_layout().

In most cases, add_subplot would be the prefered method to create axes for plots on a canvas. Only in cases where exact positioning matters, add_axes might be useful.

<b><big>2. GridSpec

```python
from matplotlib.gridspec import GridSpec
# gridspec defines a grid with a num of rows and columns
# we define later how much of the grid a subplot should span

# Constrained_layout allows some space between figures
fig = plt.figure(constrained_layout=True)
gs = GridSpec(3,3,figure=fig)

sub1 = plt.subplot(gs[0,:])
sub2 = plt.subplot(gs[1,0:2])
sub3 = plt.subplot(gs[2,0])
sub4 = plt.subplot(gs[2,1])
sub5 = plt.subplot(gs[1:3,2])
```

```python
# another way:
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3,3)

optional_params = dict(xy=(0.5,0.5),va='center',ha='center')

fig.add_subplot(gs[0,:]).annotate('1',**optional_params)
fig.add_subplot(gs[1,0:2]).annotate('2',**optional_params)
fig.add_subplot(gs[2,0]).annotate('3',**optional_params)
fig.add_subplot(gs[2,1]).annotate('4',**optional_params)
fig.add_subplot(gs[1:3,2]).annotate('5',**optional_params)
```

```python
# Using width ratios
fig = plt.figure(constrained_layout=True)

widths = [1,3,6]
heights = [1,2,4]

gs = fig.add_gridspec(3,3,width_ratios=widths,height_ratios=heights)

for row in range(3):
    for col in range(3):
        ax = fig.add_subplot(gs[row,col])
        label = 'width: {}\nheight: {}'.format(widths[col],heights[row])
        ax.text(0.1,0.2,label)   
```

```python
# subgridspecs
fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(1,2)

gs00 = gs[0].subgridspec(2,3)
gs01 = gs[1].subgridspec(3,2)

for i in range(2):
    for j in range(3):
        fig.add_subplot(gs00[i,j])
        fig.add_subplot(gs01[j,i])
```

<b><big>3. Histograms

```python
x = np.random.normal(size=500)

n, bins, patches = plt.hist(x)

# plt.hist returns a tuple of 3 objects (n,bins,patches)
# n: array of frequencies totaling the size of the sample = 500
# bins: boundaries of each patch
# patches : the histogram rectangles with their properties

print(n)
print(bins)
for i in range(len(bins)-1):
    print(patches[i])
```

```python
# we can set density=True and the hsitogram will form a probability density
# the frequencies are normalized and the integral of the histogram makes the sum of 1
n,bins,patches = plt.hist(x,bins=50,density=True)
plt.show()

print('Area: ',np.sum(n*np.diff(bins)))
```

```python
# We can also plot a histogram as a cumulative distribution function
plt.hist(x,bins=50,density=True,cumulative=True)
plt.show()
```

```python
# we can set the number of bins and the orientation
# we can also 
plt.hist(x,bins=20,orientation='horizontal',rwidth=0.7,
         edgecolor='maroon',color='red')
plt.show()
```

<b><big>4. Bar Charts

```python
# for bar plots, the width of the rectangle has no mathematical meaning
bars = plt.bar([1,2,3,4,5],
               [10,20,50,80,100],
               tick_label=['Mon','Tue','Wed','Thu','Fri'])

# set color
colors = ['lightblue','deepskyblue','skyblue','lightskyblue','steelblue']
for i in range(len(colors)):
    bars[i].set_color(colors[i])
    
plt.ylabel('Mood at work (%)')
```

```python
# Bar plots with customized ticks
from matplotlib.ticker import FuncFormatter

def millions(x,pos):
    return f'${x*1e-6:1.1f}M'
# use the user defined function for formatting
formatter = FuncFormatter(millions)

years = ('US', 'EU', 'China', 'Japan', 'Germany', 'UK', 'France', 'India')
GDP = (20494050, 18750052, 13407398, 4971929, 4000386, 2828644, 2775252, 2716746)

fig, ax = plt.subplots()

ax.yaxis.set_major_formatter(formatter)

ax.bar(x=np.arange(len(GDP)),height=GDP,tick_label=years)
ax.set_ylabel('GDP in $')

# rotate the tick labels:
plt.xticks(rotation=90)

# Invert the axis:
ax.invert_yaxis()
```

```python
# Horizontal bar chart
fig, ax = plt.subplots()

ax.barh([1,2,3,4], [1,2,3,4]) 
plt.show()
```

```python
# Grouped bar charts
# if we don't manipulate the width we will get stacked bar charts

x1 = [12,56,67,88]
x2 = [87,29,17,55]

fig, ax = plt.subplots(figsize=(4,3))

width = 0.35
ticks = np.arange(len(x1))

ax.bar(ticks,x1,width,label='x1')
ax.bar(ticks+width,x2,width,label='x2')

ax.set_xticks(ticks + width/2)

ax.legend(loc='best')
```

<b><big>5. Contour Plots

```python
# Creating a meshgrid
x_vals = np.arange(-3,4,1)
y_vals = np.arange(-3,4,1)

# return coordinate matrices from coordinate vectors
X, Y = np.meshgrid(x_vals,y_vals)

# visualize the meshgrid
fig, ax = plt.subplots()
ax.scatter(X,Y)
```

```python
# making a contour plot
Z = np.sqrt(X**2 + Y**2)

fig, ax = plt.subplots()
cs = ax.contour(X,Y,Z,linestyles='dashed')
ax.clabel(cs,fmt='%2.1f')
```

```python
# Filled contours

fig, ax = plt.subplots()
cf = ax.contourf(X,Y,Z)

plt.colorbar(cf)
plt.show()
```

```python
# we can choose the level we want to show
levels = [0.0, 0.2, 0.5, 0.9, 1.5, 2.5, 3.5]

fig = plt.figure()

cs = plt.contour(X,Y,Z,levels,colors='k')
plt.clabel(cs,fmt='%2.1f')

cf = plt.contourf(X,Y,Z,levels)
plt.colorbar()
```
